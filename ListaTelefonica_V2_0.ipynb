```python
import pandas as pd

contatos = pd.DataFrame({'Nome':['Vinicius', 'Xoris','Luh'],'Telefone':['(31) 99594-4742','(31) 96542-7854','(31) 95214-6578']})
```

```python
from IPython.display import clear_output

def limparTela():
  clear_output(wait=True)


def voltar(entrada):
  limparTela()
  print('')
  confirmacao = input("Deseja voltar? Precione ENTER para voltar")
  if confirmacao == "":
    return True
  else:
    return confirmacao

def limparFormatacao(numero):
  numeroLimpo = numero.replace("(", "").replace(")", "").replace("-", "").replace(" ", "")       # Remove quaisquer espaços, parênteses ou hífens já existentes
  return numeroLimpo

def formatarTelefone(numero):
  numeroLimpo = limparFormatacao(numero)
  while len(numeroLimpo) != 11:                                                                  # Certifica-se de que o número tem ao menos 11 dígitos (2 do DDD + 9 do telefone)
    limparTela()
    print("O número precisa ter 11 dígitos")
    numero = input('Digite o número de telefone (ddd + Número): ' )
    numeroLimpo = limparFormatacao(numero)
  numeroFormatado = numeroLimpo
  numeroFormatado = f"({numeroLimpo[:2]}) {numeroLimpo[2:7]}-{numeroLimpo[7:]}"                                 # Formata o número como (DD) xxxxx-xxxx
  if verificarDuplicado(numeroFormatado) == True:
    return -1
  else:
    return numeroFormatado


def verificarDuplicado(numero):
  global contatos
  if numero in contatos['Telefone'].values:
    return True
  else:
    return False

def ordenarContatos():
  global contatos
  contatosOrdenados = contatos.sort_values(by='Nome', key=lambda col: col.str.lower(), ignore_index = True)
  return contatosOrdenados


def exibirLista():
  limparTela()
  continuar = False
  while continuar == False:
    global contatos
    print('=================================')
    print('Contatos salvos')
    print('=================================')
    print(ordenarContatos())
    print('=================================')
    entrada = input("precione ENTER para voltar")
    if entrada == "":
      continuar = True
    else:
      limparTela()


def adicinarContato():
  limparTela()
  continuar = False
  while continuar == False:
    global contatos
    print('=================================')
    print('Adicionar contatos')
    print('=================================')
    nome = input('Digite o nome: ')
    if nome == "":
      resultado = voltar(nome)
      if resultado == True:
        continuar = True
        break
      else:
        limparTela()
        nome = resultado
    numero = input('Digite o número de telefone (ddd + Número): ' )
    numero = formatarTelefone(numero)
    if numero == -1:
      limparTela()
      print("Número já cadastrado")
      entrada = input("precione ENTER para voltar")
      if entrada == "":
        continuar = True
      else:
        limparTela()
      continuar = True
      break
    else:
      contatos.loc[len(contatos.index)] = [nome, numero]
      limparTela()
      print('Contato adicionado com sucesso')
      entrada = input("precione ENTER para voltar")
      if entrada == "":
        continuar = True
      else:
        limparTela()


def removerContato():
  limparTela()
  continuar = False
  while continuar == False:
    global contatos
    id = -1
    print('=================================')
    print('Remover contatos')
    print('=================================')
    print(ordenarContatos())
    print('=================================')
    id = input('Digite o número do contato que deseja remover:')
    if id == "":
      resultado = voltar(id)
      if resultado == True:
        continuar = True
        break
      else:
        limparTela()
        id = resultado
    contatos = ordenarContatos()
    contatos = contatos.drop(int(id))
    contatos = contatos.reset_index(drop=True)
    limparTela()
    print('Contato removido com sucesso')
    entrada = input("precione ENTER para voltar")
    if entrada == "":
      continuar = True
    else:
      limparTela()


def exportarCSV():
  global contatos
  contatosOrdenados = contatos.sort_values(by='Nome', key=lambda col: col.str.lower(), ignore_index = True)
  contatosOrdenados.to_csv('Contatos.csv', index=False)


opcao = 0

while opcao != '5':
  limparTela()
  print('=================================')
  print('Lista telefônica')
  print('=================================')
  print('1 - Exibir lista de contatos')
  print('2 - Adicionar contato')
  print('3 - Remover contato')
  print('4 - Exportar CSV')
  print('5 - Sair')
  print('=================================')

  opcao = input()
  match opcao:
    case '1':
      exibirLista()
    case '2':
      limparTela()
      adicinarContato()
    case '3':
      print(contatos)
      removerContato()
    case '4':
      print('Exportando...')
      exportarCSV()
    case '5':
      limparTela()
      print('Saindo...')
    case _:
      print('Opção inválida')


```
